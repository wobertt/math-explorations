### Problem statement:
Let $S$ be a spherical shell of radius $1$, i.e., the set of points satisfying $x^2 + y^2 + z^2 = 1$. Find the average
straight line distance between two points of $S$.

### Solution:
Fix the first point at some location.

Then the set of points $\theta$ radians away forms a circle with radius $\sin(\theta)$ and circumference $2\pi\sin(\theta)$. These points are $2\sin\left(\frac{\theta}{2}\right)$ units away from the first point.

Thus, the answer is:
$$
\begin{aligned}
&\frac{1}{4\pi}\int_0^\pi 2\pi\sin(\theta) \cdot 2\sin\left(\frac{\theta}{2}\right)d\theta \\
&= \int_0^\pi \sin\left(\frac{\theta}{2}\right) \sin(\theta) d\theta \\
&= 2\int_0^\pi \sin^2\left(\frac{\theta}{2}\right) \cos\left(\frac{\theta}{2}\right)d\theta \\
&= 2 \left[ \frac{2}{3}\sin^3\left(\frac{\theta}{2}\right)\right]_0^\pi \\
&= \frac{4}{3}
\end{aligned}
$$
